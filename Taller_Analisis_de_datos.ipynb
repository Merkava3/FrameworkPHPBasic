```python
# Creación de la figura y el eje
fig1, ax1 = plt.subplots()

# Creación del histograma de la variable 'Edad'
ax1.hist(data['Edad'], bins=30, edgecolor='black')

# Establecimiento de las etiquetas del eje
ax1.set_ylabel('Pacientes con COVID-19')
ax1.set_xlabel('Edad')

# Guardado de la figura con alta resolución
fig1.savefig("Hist_Edad.png", dpi=600)

# Mostrar la figura (opcional, si deseas verla antes de guardarla)
plt.show()
```

```python
import seaborn as sns
# Creación de la figura y el eje
fig1, ax1 = plt.subplots()

# Creación del gráfico de densidad para la variable 'Edad'
sns.kdeplot(data['Edad'], ax=ax1, fill=True)

# Establecimiento de las etiquetas del eje
ax1.set_ylabel('Densidad')
ax1.set_xlabel('Edad')

# Guardado de la figura con alta resolución
fig1.savefig("Hist_Edad.png", dpi=600)

# Mostrar la figura (opcional, si deseas verla antes de guardarla)
plt.show()
```

```python
# diagrama Box especifica de incapacidad
data['Edad'].plot(kind="box")
```

```python
# Creación de la figura y el eje
fig, ax = plt.subplots(figsize=(10, 6))

# Creación del gráfico de densidad y del diagrama de caja para la variable 'Edad'
sns.kdeplot(data['Edad'], ax=ax, fill=True, color="skyblue", alpha=0.4)
sns.boxplot(x=data['Edad'], ax=ax, color='skyblue', fliersize=5, linewidth=2)

# Añadir títulos y etiquetas
ax.set_title('Distribución y Diagrama de Caja de la Edad de los Pacientes con COVID-19')
ax.set_xlabel('Edad')
ax.set_ylabel('Densidad')

# Guardado de la figura con alta resolución
fig.savefig("KDE_BoxPlot_Edad.png", dpi=600)

# Mostrar la figura
plt.show()
```

```python
# Creación de la figura y el eje
fig, ax = plt.subplots(figsize=(10, 6))  # Ajustar el tamaño de la figura

# Creación del diagrama de caja para la variable 'Edad'
data['Edad'].plot(kind="box", ax=ax, vert=False, patch_artist=True,
                  boxprops=dict(facecolor='skyblue', color='black'),
                  whiskerprops=dict(color='black'),
                  capprops=dict(color='black'),
                  medianprops=dict(color='red'))

# Añadir títulos y etiquetas
ax.set_title('Diagrama de Caja de la Edad de los Pacientes con COVID-19')
ax.set_xlabel('Edad')
ax.set_ylabel('Pacientes')

# Añadir anotaciones para resaltar los elementos del box plot
ax.annotate('Mediana', xy=(data['Edad'].median(), 1), xytext=(data['Edad'].median() + 5, 1.1),
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='right', verticalalignment='top')

# Guardado de la figura con alta resolución
fig.savefig("BoxPlot_Edad.png", dpi=600)

# Mostrar la figura
plt.show()
```

# Decriptivos de Variable Categóricas

```python
#Conocemos las variables Categoricas : Value_count()
# Obtener el conteo de valores y generar el gráfico de pastel
ax = data['Estado'].value_counts().plot(kind='pie', autopct='%.0f%%')
# Ocultar la etiqueta del eje Y
ax.set_ylabel('')
# Establecer el título en la parte superior
ax.set_title("Distribución de  Variable Estado", pad=20, fontsize=14)
# Guardar el gráfico como una figura con 300 dpi (por ejemplo, en formato PNG)
plt.savefig('Estado.png', dpi=300)
# Mostrar el gráfico en pantalla (opcional)
plt.show()
```

```python
# prompt: diagrama de barras de estado

import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar los datos por estado
grouped_data = data.groupby('Estado').size()

# Crear el diagrama de barras
sns.barplot(x=grouped_data.index, y=grouped_data.values)

# Configurar el título y las etiquetas
plt.title('Distribución de Pacientes por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Pacientes')

# Rotar las etiquetas del eje X para mejorar la legibilidad
plt.xticks(rotation=45)

# Guardar el gráfico como una figura con 300 dpi (por ejemplo, en formato PNG)
plt.savefig('Diagrama_Barras_Estado.png', dpi=300)

# Mostrar el gráfico en pantalla (opcional)
plt.show()
```
